```python
# Cell #2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

# The line below will generate an error unless you set this environment variable in your shell prior to launching jupyter lab:
# export DRJIT_LIBLLVM_PATH=/absolute/path/to/libLLVM.dylib

from torchrf import rt
```

```python
# Cell #3
scene = rt.load_scene(rt.scene.munich)
```

```python
# Cell #4
# The Sionna tutorial has a special mode for Google Colab that is not implemented here.
scene.preview()
```

```python
# Cell #5
scene.render(camera="preview", num_samples=512)
```

```python
# Cell #6
render_to_file = True # Set to True to render image to file

# Render scene to file from preview viewpoint
if render_to_file:
    scene.render_to_file(camera="scene-cam-0", # Also try camera="preview"
                         filename="scene.png",
                         resolution=[650,500])
```
